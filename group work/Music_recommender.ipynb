# Build Recommender System

### Based on the features (user_id, track_name, artist_name) available in the dataset, the collaborative filtering algorithm will be employed as the dataset does not contain comprehensive features.

```python
# import modules
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
```

```python
# save to csv
df_tsv_filled.to_csv('music_preprocessed.csv')
```

```python
# Set the chunk size to help solve memory error issues
chunk_size = 100000

# Initialize an empty list to store the processed chunks
processed_chunks = []

# Read the data in chunks using Pandas read_csv
for chunk in pd.read_csv('music_preprocessed.csv', chunksize=chunk_size):
    # Perform processing on the chunk (e.g., filtering, aggregation, etc.)
    interaction_matrix_chunk = pd.pivot_table(chunk, index='user_id', columns='track_name', aggfunc=lambda x: 1, fill_value=0)
    
    # Append the processed chunk to the list
    processed_chunks.append(interaction_matrix_chunk)

# Combine the results from all chunks
interaction_matrix = pd.concat(processed_chunks)

# view full interaction matrix
interaction_matrix.head(20)
```

# User Similarity Matrix usine cosine similarity

```python
# Transpose the interaction matrix to get item-user matrix for similarity calculation
item_user_matrix = interaction_matrix.T

# Compute the cosine similarity matrix
user_similarity = cosine_similarity(item_user_matrix)

# Convert the similarity matrix to a DataFrame
user_similarity_df = pd.DataFrame(user_similarity, index=interaction_matrix.columns, columns=interaction_matrix.columns)

# Check the user similarity matrix
user_similarity_df.head()
```

# User-based collaborative filtering function

```python
def user_based_collaborative_filtering(user_id, top_n=5):
    # Get the similarity values for the given user
    user_similarity_values = user_similarity_df.loc[user_id]

    # Find the users most similar to the given user
    most_similar_users = user_similarity_values.sort_values(ascending=False)[1:top_n+1]

    # Get the interaction matrix for the similar users
    similar_users_interactions = interaction_matrix.loc[most_similar_users.index]

    # Sum the interaction scores for each track, weighted by similarity
    track_scores = similar_users_interactions.apply(lambda row: row * user_similarity_values[row.name], axis=1).sum()

    # Sort the tracks by score in descending order
    recommended_tracks = track_scores.sort_values(ascending=False)

    return recommended_tracks.index.tolist()
```

```python
# Generate Recommendations for User

user_id_to_recommend = 'user_id'
top_n_recommendations = 5

recommended_tracks = user_based_collaborative_filtering(user_id_to_recommend, top_n=top_n_recommendations)

print("Recommended tracks for User ID {}: \n{}".format(user_id_to_recommend, recommended_tracks))
```
